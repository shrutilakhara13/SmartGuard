```python
def data_cleaning(df):
    # Remove leading and trailing whitespaces from column names
    df.columns = df.columns.str.strip()
    print("Dataset Shape:", df.shape)

    # Replace negative numeric values with 0
    num = df._get_numeric_data()
    num[num < 0] = 0

    # Identify and drop columns with zero variance
    zero_variance_cols = []
    for col in df.columns:
        if len(df[col].unique()) == 1:
            zero_variance_cols.append(col)
    df.drop(columns=zero_variance_cols, axis=1, inplace=True)
    print("Zero Variance Columns:", zero_variance_cols, "are dropped!!")
    print("Shape after removing the zero variance columns:", df.shape)

    # Replace infinite values with NaN and drop rows with NaN
    df.replace([np.inf, -np.inf], np.nan, inplace=True)
    print(df.isna().any(axis=1).sum(), "rows dropped")
    df.dropna(inplace=True)
    print("Shape after Removing NaN:", df.shape)

    # Drop duplicate rows
    df.drop_duplicates(inplace=True)
    print("Shape after dropping duplicates:", df.shape)

    # Identify and drop columns with identical values
    column_pairs = [(i, j) for i, j in combinations(df, 2) if df[i].equals(df[j])]
    identical_cols = []
    for col_pair in column_pairs:
        identical_cols.append(col_pair[1])
    df.drop(columns=identical_cols, axis=1, inplace=True)
    print("Columns which have identical values:", column_pairs, "dropped!")
    print("Shape after removing identical value columns:", df.shape)

    return df
```

<h3>Balancing the dataset</h3>

```python
# Calculate the size of the subset where the label is 'ATTACK'
size = len(df.loc[df.label == 'ATTACK'])
print("Size of 'ATTACK' subset:", size)

# Create a balanced DataFrame by sampling the same number of rows for each label
# The lambda function is used with groupby to apply the sampling to each group (label)
bal_df = df.groupby('label').apply(lambda x: x.sample(n=min(size, len(x))))

# The resulting DataFrame 'bal_df' now has an equal number of rows for each label
```

```python
# Import necessary modules for data preprocessing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

# Extract features (X) and labels (y) from the balanced DataFrame 'bal_df'
X = bal_df.drop(columns='label')  # Features: Exclude the 'label' column
y = bal_df['label'].astype('int')  # Labels: Convert to integer format for classification

# Apply Min-Max scaling to normalize feature values between 0 and 1
X = MinMaxScaler().fit_transform(X)
```

```python
# Split the data into training and testing sets using train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=None)

# Display the shapes of the training and testing sets
print("Training set shapes - X_train:", X_train.shape, "y_train:", y_train.shape)
print("Testing set shapes - X_test:", X_test.shape, "y_test:", y_test.shape)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, accuracy_score, f1_score, precision_score, recall_score
def classify(model):
    # Fit the model on the training data
    model.fit(X_train, y_train)
    # Evaluate and print the accuracy on the test set
    model_score = model.score(X_test, y_test)
    print(f"Model Accuracy on Test Set: {model_score:.4f}")

    # Make predictions on the test set
    y_pred = model.predict(X_test)
    # Print the classification report
    print("Classification Report:")
    print(classification_report(y_test, y_pred))
    # Create a figure for visualizations
    fig = plt.figure(figsize=(10, 5))
    # Plot the confusion matrix
    plt.subplot(1, 2, 1)
    cm = confusion_matrix(y_test, y_pred, normalize='true')
    sns.heatmap(cm, annot=True)
    
    # Plot the ROC curve
    plt.subplot(1, 2, 2)
    fpr, tpr, thresholds = roc_curve(y_test, y_pred)
    plt.plot(fpr, tpr, label='ROC Curve')
    plt.plot([0, 1], [1, 0], 'k--', label='Random Guess')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.xlim([-0.02, 1])
    plt.ylim([0, 1.02])
    plt.legend(loc='lower right')

    # Print additional metrics
    print("Additional Metrics:")
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print("Precision:", precision_score(y_test, y_pred))
    print("Recall:", recall_score(y_test, y_pred))
    print("F1 Score:", f1_score(y_test, y_pred))
```

```python
def classify_unbalanced(model):
    # Fit the unbalanced model on the training data
    model.fit(Xu_train, yu_train)
    
    # Make predictions on the test set
    y_pred = model.predict(Xu_test)
    
    # Print the classification report
    print("Classification Report:")
    print(classification_report(yu_test, y_pred))

    # Create a figure for visualizations
    fig = plt.figure(figsize=(10, 5))

    # Plot the confusion matrix
    plt.subplot(1, 2, 1)
    cm = confusion_matrix(yu_test, y_pred, normalize='true')
    sns.heatmap(cm, annot=True)

    # Plot the ROC curve
    plt.subplot(1, 2, 2)
    fpr, tpr, thresholds = roc_curve(yu_test, y_pred)
    plt.plot(fpr, tpr, label='ROC Curve')
    plt.plot([0, 1], [1, 0], 'k--', label='Random Guess')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.xlim([-0.02, 1])
    plt.ylim([0, 1.02])
    plt.legend(loc='lower right')

    # Print additional metrics
    print("Additional Metrics:")
    print("Accuracy:", accuracy_score(yu_test, y_pred))
    print("Precision:", precision_score(yu_test, y_pred))
    print("Recall:", recall_score(yu_test, y_pred))
    print("F1 Score:", f1_score(yu_test, y_pred))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

def classify_with_lstm(X_train, y_train, X_test, y_test):
    # Reshape the data for LSTM input
    X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
    X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

    # Define the LSTM model
    model = Sequential()
    model.add(LSTM(50, input_shape=(X_train.shape[1], 1)))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

    # Train the LSTM model
    model.fit(X_train, y_train, epochs=5, batch_size=32, validation_split=0.2)

    # Make predictions on the test set
    y_pred = (model.predict(X_test) > 0.5).astype("int32")

    # Print classification report and confusion matrix
    print(classification_report(y_test, y_pred))
    cm = confusion_matrix(y_test, y_pred, normalize='true')
    sns.heatmap(cm, annot=True, fmt=".2f")

    # Additional metrics
    print("The Accuracy of the Model is", accuracy_score(y_test, y_pred))
    print("The Precision of the Model is", precision_score(y_test, y_pred))
    print("The Recall of the Model is", recall_score(y_test, y_pred))
    print("The F1 Score of the Model is", f1_score(y_test, y_pred))
```

```python
classify_with_lstm(X_train, y_train, X_test, y_test)
```
